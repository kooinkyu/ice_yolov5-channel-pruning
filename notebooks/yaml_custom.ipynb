```python
import torch, time
from thop import profile
from models.common import DetectMultiBackend
from utils.torch_utils import select_device

weights = 'runs/train/ice_eye_mouth20_pruned_finetune_25_30/weights/best.pt'
device  = select_device('0')

model = DetectMultiBackend(weights, device=device)
model.eval()

# ğŸ”¹ Params
params = sum(p.numel() for p in model.model.parameters())
print(f"Parameters: {params/1e6:.2f} M")

# ğŸ”¹ GFLOPs (thop)
dummy = torch.randn(1, 3, 640, 640).to(device)
flops, _ = profile(model.model, inputs=(dummy,), verbose=False)
print(f"GFLOPs: {flops/1e9:.2f}")

# ğŸ”¹ Latency & FPS
torch.cuda.synchronize()
t0 = time.time()
runs = 100
for _ in range(runs):
    _ = model(dummy)
torch.cuda.synchronize()
t1 = time.time()
latency = (t1 - t0) / runs * 1000  # ms/image
print(f"Latency: {latency:.2f} ms | FPS: {1000/latency:.2f}")
```
